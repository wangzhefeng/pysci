# Matplotlib 与 Seaborn 对比

## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("classic")

# data
rng = np.random.RandomState(42)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), axis = 0)

# plot
plt.plot(x, y)
plt.legend("ABCDEF", ncol = 2, loc = "best");
```

## Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# data
rng = np.random.RandomState(42)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), axis = 0)

# plot
plt.plot(x, y)
plt.legend("ABCDEF", ncol = 2, loc = "best");
```

# 频次直方图

## matplotlib 

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
plt.style.use("classic")
for col in "xy":
    plt.hist(x = data[col], alpha = 0.5);
```

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
sns.set()
for col in "xy":
    plt.hist(x = data[col], alpha = 0.5);
```

## seaborn

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
for col in "xy":
    sns.histplot(data = data, x = col, alpha = 0.5);
```

# KDE

## kde

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
for col in "xy":
    sns.kdeplot(data = data, x = col, fill = True);
```

## kde & hist

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
sns.histplot(data = data, x = "x", kde = True)
sns.histplot(data = data, x = "y", kde = True);
```

# 密度图

## 二维 kde

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
sns.kdeplot(data = data, x = "x", y = "y");
```

## 二维 kde 的联合分布图

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
with sns.axes_style("white"):
    sns.jointplot(data = data, x = "x", y = "y", kind = "kde");
```

```python
# data
data = np.random.multivariate_normal(
    mean = [0, 0],
    cov = [[5, 2], [2, 2]],
    size = 2000,
)
data = pd.DataFrame(data, columns = ["x", "y"])

# plot
with sns.axes_style("white"):
    sns.jointplot(data = data, x = "x", y = "y", kind = "hex");
```

# 分面频次直方图

```python
# data
tips = sns.load_dataset("tips")
tips["tip_pct"] = 100 * tips["tip"] / tips["total_bill"]
tips
```

```python
# plot
grid = sns.FacetGrid(data = tips, row = "sex", col = "time", margin_titles = True)
grid.map(plt.hist, "tip_pct", bins = np.linspace(0, 40, 15));
```

# 因子图

```python
with sns.axes_style(style = "ticks"):
    g = sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex", kind = "box")
    g.set_axis_labels("Day", "Total Bill");
```

# 联合分布

```python
with sns.axes_style("white"):
    sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "hex")
```

```python
with sns.axes_style("white"):
    sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "reg")
```

# 条形图

```python
# data
planets = sns.load_dataset("planets")
planets
```

```python
with sns.axes_style("white"):
    g = sns.catplot(data = planets, x = "year", aspect = 2, kind = "count", color = "steelblue")
    g.set_xticklabels(step = 5);
```

```python
with sns.axes_style("white"):
    g = sns.catplot(
        data = planets, 
        x = "year", 
        aspect = 4.0, 
        kind = "count", 
        hue = "method", 
        order = range(2001, 2015)
    )
    g.set_ylabels("Number of Planets Discovered");
```
