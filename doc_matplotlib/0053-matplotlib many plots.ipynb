```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

# Bar

### Stacked Bar Graph

```python
# data
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(5)
width = 0.35

# plot
fig = plt.figure()

p1 = plt.bar(x = ind, height = menMeans, width = width, 
             yerr = menStd)
p2 = plt.bar(x = ind, height = womenMeans, width = width, 
             bottom = menMeans, 
             yerr = womenStd)

plt.title("Scores by group and gender")
plt.xlabel('group')
plt.ylabel('Scores')
plt.xlim()
plt.ylim(0, 81)
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

# plt.draw()
# plt.show()
```

### Horizontal bar chart

```python
#  data
people = np.array(['Tom', 'Dick', 'Harry', 'Slim', 'Jim'])
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

# config
plt.rcdefaults()
fig, ax = plt.subplots()

# plot
ax.barh(y = y_pos, width = performance, 
        xerr = error, align = 'center', 
        color = 'green', ecolor = 'black')

ax.set_title('How fast do you want to go today?')
ax.set_xlabel('Performance')
ax.set_ylabel('People')
# ax.set_ylim()
# ax.set_xlim()
# ax.set_xticks([])
ax.set_yticks(y_pos)
# ax.set_xticklabels([])
ax.set_yticklabels(people)

plt.show()
```

### Bar plot with errorbars and height labels

```python
# data
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(len(menMeans))
group = ('G1', 'G2', 'G3', 'G4', 'G5')
width = 0.35

# config
fig, ax = plt.subplots()


men_xloc = ind - width / 2
women_xloc = ind + width / 2

# plot
rects1 = ax.bar(x = men_xloc, height = menMeans, width = width, 
                yerr = menStd, color = 'SkyBlue',
                label = 'Men')
rects2 = ax.bar(x = women_xloc, height = womenMeans, width = width,
                yerr = womenStd, color = 'IndianRed', 
                label = 'Women')

ax.set_title('Scoresby group and gender')
ax.set_xlabel('Group')
ax.set_ylabel('Scores')
ax.set_xticks(ind)
ax.set_xticklabels(group)
ax.set_yticks(np.arange(0, 40, 5))
# as.set_yticklabels()

def autolable(rects, xpos = 'center'):
    xpos = xpos.lower()
    ha = {
        'center': 'center',
        'right': 'left',
        'left': 'right'
    }
    offset = {
        'left': 0.43,
        'center': 0.5,
        'right': 0.57,
    }
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() * offset[xpos],
                1.01 * height,
                '{}'.format(height),
                ha = ha[xpos], 
                va = 'bottom')

autolable(rects1, 'left')
autolable(rects2, 'right')

plt.show()
```
