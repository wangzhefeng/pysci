# 1.APIs: `matplotlib.pyplot.subplots()`

### data

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

### 1.1 fig, ax = plt.subplots()

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("A single plot")
```

### 1.2 fig, ax = plt.subplots(nrows, ncols)

```python
fig, ax = plt.subplots(nrows = 2)
ax[0].plot(x, y)
ax[1].plot(x, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2)
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Axis [0, 0]")

axs[0, 1].plot(x, y, "tab:orange")
axs[0, 1].set_title("Axis [0, 1]")

axs[1, 0].plot(x, -y, "tab:green")
axs[1, 0].set_title("Axis [1, 0]")

axs[1, 1].plot(x, -y, "tab:red")
axs[1, 1].set_title("Axis [1, 1]")

for ax in axs.flat:
    ax.set(xlabel = "x-label", ylabel = "y-label")
    ax.label_outer()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2)
ax1.plot(x, y)
ax2.plot(x, y ** 2, "tab:orange")
ax3.plot(x, -y, "tab:green")
ax4.plot(x, -y ** 2, "tab:red")

for ax in fig.get_axes():
    ax.label_outer()
```

### 1.3 fig, axs = plt.subplots(nrows, ncols, sharex, sharey)

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2)
ax1.plot(x, y)
ax2.plot(x + 1, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2, sharex = True)
ax1.plot(x, y)
ax2.plot(x + 1, -y)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, sharex = True, sharey = True)
ax1.plot(x, y ** 2)
ax2.plot(x, 0.3 * y, 'o')
ax3.plot(x, y, "+")
```

### 1.4 fig, axs = plt.subplots(nrows, ncols, sharex, sharey, gridspec_kw = {"hspace": 0, "wspace": 0})

```python
fig, axs = plt.subplots(nrows = 3, ncols = 1, 
                        sharex = True, sharey = True, 
                        gridspec_kw = {"hspace": 0})
axs[0].plot(x, y ** 2)
axs[1].plot(x, 0.3 * y, 'o')
axs[2].plot(x, y, '+')

for ax in axs:
    ax.label_outer()
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, 
                        sharex = "col", sharey = "row", 
                        gridspec_kw = {"hspace": 0, "wspace": 0})
(ax1, ax2), (ax3, ax4) = axs
ax1.plot(x, y)
ax2.plot(x, y ** 2, "tab:orange")
ax3.plot(x + 1, -y, "tab:green")
ax4.plot(x + 2, -y ** 2, "tab:red")

for ax in axs.flat:
    ax.label_outer()
```

### 1.5 fig, axs = plt.subplots(nrows, ncols, subplot_kw = dict(projection = "polar"))

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, 
                               subplot_kw = dict(projection = "polar"))
ax1.plot(x, y)
ax2.plot(x, y ** 2)
```
