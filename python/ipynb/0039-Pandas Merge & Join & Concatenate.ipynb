# 1.Concatenating

````python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index = [0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
````

````python
df1
````

````python
df2
````

````python
df3
````

````python
frames = [df1, df2, df3]
result = pd.concat(frames)
result
````

## 1.1 `pd.concat()`

```python
pd.concat(objs, axis = 0, join = "outer", join_axes = None, ignore_index = False, 
          keys= None, levles = None, names = None, verify_integrity = False, 
          copy = True)
```

* union
    - `pd.concat(objs, axis = 0, keys = [])`
* outer full join 
    - `pd.concat(objs, axis = 1, join = "outer", sort = False)`
* inner join
    - `pd.concat(objs, axis = 1, join = "inner", sort = False)`
* left join
    - `pd.concat(objs, axis = 1, join = "outer", join_axes = [])`
* right join
    - `pd.concat(objs, axis = 1, join = "outer", join_axes = [])`


### 1.1.1 `pd.concat(objs, axis = 0, keys = [])`

````python
result = pd.concat(frames, axis = 0, keys = ["x", "y", "z"])
result
````

````python
result.loc["x"]
````

### 1.1.2 `pd.concat(objs, axis = 1, join = "outer", sort = False)`

````python
df1
````

````python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
df4
````

````python
result = pd.concat([df1, df4], axis = 1, join = "outer", sort = False)
result
````

### 1.1.3 `pd.concat(objs, axis = 1, join = "inner", sort = False)`

````python
result = pd.concat([df1, df4], axis = 1, join = "inner", sort = False)
result
````

### 1.1.4 `pd.concat(objs, axis = 1, join = "inner", join_axes = [df1.index])`

````python
result = pd.concat([df1, df4], axis = 1, join_axes = [df1.index])
result
````
