#### (8) 最近异常的类型,值,追踪对象

```python
try:
    raise IndexError
except:
    print(sys.exc_info())
```

```python
import traceback, sys

def grail(x):
    raise TypeError("Already got one")

try:
    grail("authur")
except:
    exc_info = sys.exc_info()
    print(exc_info)
    print(exc_info[0])
    print(exc_info[1])
    print(exc_info[2])
    traceback.print_tb(exc_info[2])
```
