# 1.Controlling the figure of seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### 绘制 $\sin$ 函数

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)

sinplot()
```

## 1.2 matplotlib 参数在 seaborn 中的分类

* 用以设置美学风格(set the aesthetic style of plot)
    - `axes_style()`
        - 返回一个参数的字典对象
    - `set_style()`
        - 设置为 matplotlib 的默认设置
* 用以按比例绘制图形的各种元素(scales various elements of the figure)
    - `plotting_context()`
        - 返回一个参数的字典对象
    - `set_context()`
        - - 设置为 matplotlib 的默认设置

#### 1.2.1.1 `sns.set_style()` 设置图形整体风格

```python
?sns.set_style
```

```python
sns.set_style("darkgrid")
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data);
```

```python
sns.set_style("whitegrid")
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data);
```

```python
sns.set_style("dark")
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data);
```

```python
sns.set_style("white")
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data);
```

```python
sns.set_style("ticks")
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data);
```

#### 1.2.1.2 `sns.despine()` 设置图像坐标轴线

```python
sinplot();
```

```python
sinplot()
sns.despine();
```

```python
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
fig, ax = plt.subplots()
sns.violinplot(data = data)
sns.despine(offset = 10, trim = True);
```

```python
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data)
sns.despine(left = True);
```

#### 1.2.1.3 sns.axes_style(style = "") 设置临时参数

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
with sns.axes_style("darkgrid"):
    ax[1] = sinplot()
ax[2] = sinplot()
```
