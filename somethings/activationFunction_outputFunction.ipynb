```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

## Activation Function

### Step function

```python
def step_function(x):
    return np.array(x > 0, dtype = np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### Sigmoid function

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### ReLU function

```python
def relu(x):
    y = np.maximum(0, x)
    return y

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(- 1, 6)
plt.show()
```

### step, sigmoid, relu function

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)
y3 = relu(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.ylim(-0.25, 1.25)
plt.show()
```

### Softmax function

```python
def softmax(a):
	exp_a = np.exp(a)
	sum_exp_a = np.sum(exp_a)
	y = exp_a / sum_exp_a
	
	return y
```
